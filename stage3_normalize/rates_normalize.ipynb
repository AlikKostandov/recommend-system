```python
import pandas as pd

rates = pd.read_csv(
    "../ml-1m/ratings.dat",
    sep="::",
    engine="python",
    names=["user_id", "movie_id", "rating", "timestamp"],
    encoding="latin-1"
)

rates.head()
```

```python
users = pd.read_csv("../users.csv")
users.head()
```

```python
movies = pd.read_csv("../movies.csv")
movies.head()
```

```python
valid_user_ids = set(users["user_id"])
valid_movie_ids = set(movies["movie_id"])

print("Before:" + str(len(rates)))
rates_clean = rates[
    rates["user_id"].isin(valid_user_ids) &
    rates["movie_id"].isin(valid_movie_ids)
].copy()
print("After:" + str(len(rates_clean)))
```

```python
rates_clean["datetime"] = pd.to_datetime(
    rates_clean["timestamp"],
    unit="s",
    utc=True
)
rates_clean.drop
```

```python
rates_clean = rates_clean.drop("timestamp", axis=1)
```

```python
rates_clean.to_csv("../user_movie_rates.csv", index=False, encoding="utf-8")
```
